```python
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
sample_md = pd.read_csv('combined-map.tsv', sep='\t', index_col=0, dtype=object)
minimum_threshold = 5721
```

```python
sample_md
```

```python
group = 'autism'
sample_type = 'stool'

def get_participant_table(sample_md, group, sample_type):
    group_md = sample_md[sample_md['Group'] == group]
    group_md = group_md[group_md['SampleType'] == sample_type]
    group_md = group_md[group_md['week'] != 'repeat']
    weeks = group_md['week'].unique()
    subjects = group_md['SubjectID'].unique()

    idx = pd.Index(zip(group_md['SubjectID'], group_md['week']))
    group_md.index = idx

    data = np.zeros((len(subjects), len(weeks)))
    for i, subject in enumerate(subjects):
        for j, week in enumerate(weeks):
            e = (subject, str(week))
            try:
                value = group_md['sequences.per.sample'][e]
            except KeyError:
                value = np.nan
            try:
                data[i, j] = value
            except ValueError:
                data[i, j] = np.max(value)
    result = pd.DataFrame(data, index=pd.Index(subjects, name='SubjectID'), columns=pd.Index(map(int, weeks), name='Week'))
    result = result.sort_index(1)
    return result
```

```python
df = get_participant_table(sample_md, 'autism', 'stool')
df
```

```python
mask = df < minimum_threshold
with sns.axes_style("white"):
    sns.heatmap(df, mask=mask)
```

```python
df = get_participant_table(sample_md, 'autism', 'swab')
df
```

```python
mask = df < minimum_threshold
with sns.axes_style("white"):
    sns.heatmap(df, mask=mask)
```

```python
mask = df < minimum_threshold
with sns.axes_style("white"):
    sns.heatmap(df, mask=mask, cmap='Greens')
```
